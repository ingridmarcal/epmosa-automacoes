```python


# Create a StringArray
string_array = np.array(names_list, dtype=str)

# Concatenate the two arrays horizontally
result = np.concatenate((string_array.reshape(-1,1), new_arr.reshape(-1,1)), axis=1)

# Remove quotation marks from strings
result[:, 0] = np.char.strip(result[:, 0], chars="'\"")

# Save result to file with .txt extension
np.savetxt('output.txt', result, delimiter=';', fmt='%s')

# Print the result
print(result)

```

```python
len(result)
```

```python
# dictionary to store unique classes and their probabilities
class_dict = {}

# iterate over original list
for item in result:
    class_name = item[0]
    prob = float(item[1])/100

    # check if class already exists in dictionary
    if class_name in class_dict:
        # calculate new probability using formula
        old_prob = class_dict[class_name]
        new_prob = 1 - (1 - old_prob) * (1 - prob)
        class_dict[class_name] = new_prob
    else:
        class_dict[class_name] = prob

# create new list with unique classes and their probabilities
new_list = [[class_name, str(60+ int(round(prob * 100, 1)))] for class_name, prob in class_dict.items()]

# print new list
for item in new_list:
    print(item)

len(new_list)
```

```python
# New list to store modified names
new_names_list = []

# Loop through each name in the original list
for name, number in new_list:
    # Find the index of the "org" substring starting from the end of the string
    org_index = name.rfind("org")

    # Get the substring starting from "org" and replace slashes with dots
    new_name = name[org_index:].replace("/", ".")

    # Add the modified name and number to the new list
    new_names_list.append([new_name, number])

# Print the new list
new_names_list
```
